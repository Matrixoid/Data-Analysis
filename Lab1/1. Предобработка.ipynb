```python
plt.figure(figsize = (30,7))
plt.plot(user_dataset['mean_score'])
plt.ylabel('mean_score')
plt.xlabel('anime site user number in dataset')
```

```python
plt.figure(figsize = (15,10))
plt.ylabel('Number of joined users')
plt.xlabel('Year of join')
plt.hist(user_dataset['joined'].apply(lambda x : float(x.split('-')[0])))
```

```python
genre_list = [i.split(': ')[0].strip() for i in (user_dataset['genre_count'][0][1:-1].split(','))]
genre_cnt = [0]*len(genre_list)
for i in range(len(genre_list)):
    g = [int(k.split(': ')[1].strip()) for k in (user_dataset['genre_count'][i][1:-1].split(','))]
    for j in range(len(g)):
        genre_cnt[j] += g[j]

plt.figure(figsize = (15, 10))
plt.hist(genre_list, len(genre_list), weights = genre_cnt, orientation = 'horizontal')
plt.xlabel('number of anime watched')
plt.ylabel('anime genres')
plt.show()
```

```python
compl = [float(i) for i in user_dataset['completed']]
day_wat = [float(i) for i in user_dataset['days_watched']]

res = [compl[i] / day_wat[i] for i in range(len(compl))]

plt.figure(figsize = (15, 20))
plt.xlabel('completed')
plt.ylabel('days_watched')
plt.hist2d(compl, day_wat, bins = 100)
plt.show()
```

```python
nba_dataset = pd.read_csv('nba2k20-full.csv')
nba_dataset
```

```python
plt.figure(figsize = (15,3))
plt.xlabel('Part of max salary')

salary = nba_dataset.salary.apply(lambda x : int(x[1:]))
max_salary = max(salary)

print(' : ' + str(max_salary))

plt.boxplot(salary.apply(lambda x : x /max_salary), vert=False, whis=0.95)
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.xlabel('weight, kg')
plt.ylabel('height, m')
plt.scatter(nba_dataset.weight.apply(lambda x : float(x.split('/')[1].split()[0])), 
            nba_dataset.height.apply(lambda x : float(x.split('/')[1])))
plt.show()
```

```python
#def years_online():
year_of_registration = user_dataset.joined.apply(lambda x : int(x.split('-')[0]))
current_year = 2021
years_online = [current_year - year_of_registration[i] for i in range(len(year_of_registration))]
user_dataset.loc[:, 'years_online'] = years_online
user_dataset.to_csv('./user_dataset.csv')
```
