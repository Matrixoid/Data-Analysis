```python
import matplotlib.pyplot as plt
from random import random as rand
```

```python
X = []
Y = []

for i in range(10):
    x = rand() * 20
    y = rand() * 20
    X.append(x)
    Y.append(y)
    
X.sort()
Y.sort()
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
from scipy import interpolate
import numpy as np

f = interpolate.interp1d(X, Y)
newX = np.linspace(min(X), max(X), 1000)
plt.plot(X, Y, "o", newX, f(newX), "--")
plt.show()
```

```python
def l_i(x, x_i):
    productnum = 1
    productden = 1
    for x_j in X:
        if (x_j != x_i):
            productnum *= (x - x_j)
    for x_j in X:
        if (x_j != x_i):
            productden *= (x_i - x_j)
    return productnum / productden 

def lagrange_func(x):
    l = 0
    for i in range(len(Y)):
        l += Y[i] * l_i(x, X[i])
    return l

def lagrange(x, y):
    return lagrange_func
```

```python
f1 = lagrange(X, Y)
f2 = interpolate.lagrange(X, Y)
newX = np.linspace(min(X), max(X), 1000)
plt.plot(X, Y, 'o', newX, f1(newX), '-', newX, f2(newX), '--')
plt.show()
```
